```python
# Выбор признаков для обучения 1 скважина
y_tr1 = np.array(data_1['Rate of Penetration m/h'])
data_1_train = data_1.drop(data_1[['anomaly']],axis=1)
x_tr1 = np.array(data_1_train)

# Выбор признаков для обучения 2 скважина
y_tr2 = np.array(data_2['Rate of Penetration m/h'])
data_2_train = data_2.drop(data_2[['anomaly']],axis=1)
x_tr2 = np.array(data_2_train)

# Выбор признаков для обучения 3 скважина
y_tr3 = np.array(data_3['Rate of Penetration m/h'])
data_3_train = data_3.drop(data_3[['anomaly']],axis=1)
x_tr3 = np.array(data_3_train)

# Выбор признаков для обучения 4 скважина
y_tr4 = np.array(data_4['Rate of Penetration m/h'])
data_4_train = data_4.drop(data_4[['anomaly']],axis=1)
x_tr4 = np.array(data_4_train)
```

```python
def DNN_model():
    
    NN_model = tf.keras.models.Sequential()

    # Входной слой :
    NN_model.add(tf.keras.layers.Dense(128, kernel_initializer='normal',input_dim =(12), activation='relu'))

    # Скрытые слои :
    NN_model.add(tf.keras.layers.Dense(256, kernel_initializer='normal',activation='relu'))
    NN_model.add(tf.keras.layers.Dense(256, kernel_initializer='normal',activation='relu'))
    NN_model.add(tf.keras.layers.Dense(256, kernel_initializer='normal',activation='relu'))

    # Выходной слой :
    NN_model.add(tf.keras.layers.Dense(1, kernel_initializer='normal',activation='linear'))

    # Компилирование модели :
    NN_model.compile(loss='mean_absolute_error', optimizer='adam', metrics="RootMeanSquaredError")
    return NN_model
model3 = DNN_model()
model3.summary()
```

```python
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=3)
```

```python
# шкалирование данных
x_train1,y_train1,x_scaler1,y_scaler1 = functions.data_encoding(x_tr1,y_tr1)
x_train2,y_train2,x_scaler2,y_scaler2 = functions.data_encoding(x_tr2,y_tr2)
x_train3,y_train3,x_scaler3,y_scaler3 = functions.data_encoding(x_tr3,y_tr3)
x_train4,y_train4,x_scaler4,y_scaler4 = functions.data_encoding(x_tr4,y_tr4)
```

```python
model3.fit(x_train1, y_train1, epochs=100, batch_size=10, validation_split = 0.2, callbacks=callback)
model3.fit(x_train2, y_train2, epochs=100, batch_size=10, validation_split = 0.2, callbacks=callback,shuffle=None)
model3.fit(x_train3, y_train3, epochs=100, batch_size=10, validation_split = 0.2, callbacks=callback,shuffle=None)
model3.fit(x_train4, y_train4, epochs=100, batch_size=10, validation_split = 0.2, callbacks=callback,shuffle=None)
```
